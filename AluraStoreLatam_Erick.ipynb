```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap

t1 = pd.read_csv(url)
t2 = pd.read_csv(url2)
t3 = pd.read_csv(url3)
t4 = pd.read_csv(url4)

t1["Tienda"] = "Tienda 1"
t2["Tienda"] = "Tienda 2"
t3["Tienda"] = "Tienda 3"
t4["Tienda"] = "Tienda 4"

df = pd.concat([t1, t2, t3, t4], ignore_index=True)

df.info()
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="lon",
    y="lat",
    hue="Tienda",
    alpha=0.6,
    palette="tab10"
)
plt.title("Distribución geográfica de las ventas por tienda", fontsize=14)
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend(title="Tienda")
plt.show()
```

```python
# Crear mapa base
m = folium.Map(
    location=[df["lat"].mean(), df["lon"].mean()],
    zoom_start=5,
    tiles="CartoDB positron"
)

# Datos para el Heatmap (ponderado por precio)
heat_data = df[["lat", "lon", "Precio"]].dropna().values.tolist()

# Añadir capa Heatmap
HeatMap(heat_data, radius=8, blur=12, max_zoom=10).add_to(m)

m
```

```python
# Crear "regiones" agrupando coordenadas aproximadas
df["region"] = list(zip(df["lat"].round(1), df["lon"].round(1)))

# Calcular métricas promedio por tienda y región
region_perf = df.groupby(["Tienda", "region"]).agg({
    "Precio": "mean",
    "Calificación": "mean"
}).reset_index()

# Visualizar el valor promedio de venta por tienda y región
plt.figure(figsize=(10, 6))
sns.boxplot(data=region_perf, x="Tienda", y="Precio", palette="Set2")
plt.title("Distribución del precio promedio por tienda y región", fontsize=14)
plt.xlabel("Tienda")
plt.ylabel("Precio promedio")
plt.show()
```
